```python
from matplotlib import pyplot as pyplt
from matplotlib import style
from matplotlib import colors
from scipy.stats import skew
import pandas as pd
from scipy import stats

path = "https://github.com/Xatri007git/My-First-Data-Visualization/blob/main/meow-total-randomised-noise.csv?raw=true"
df = pd.read_csv(path)

rows = df.shape[0] # stores the number of rows
cols = df.shape[1] # stores the number of columns

columns = list(df.columns) #stores the column names of the unmodified dataframe

def replace_woof():
    for i in range(len(columns)):
     df.loc[df[columns[i]] == "woof", columns[i]] = 0

def Best_4():
   
 Tuts = ["T1","T2","T3","T4","T5"]
 Best_4_Tuts = []
 worst_tuts = []

 for i in range(rows):
    worst_tut = "T1"
    for tut in Tuts:
       if(int(df[tut][i]) < int(df[worst_tut][i])):
          worst_tut = tut
    worst_tuts.append(worst_tut)

 for i in range(rows):
    best_4 = ""
    for tut in Tuts:
       if(tut != worst_tuts[i]):
          best_4 += tut + " "
    Best_4_Tuts.append(best_4)

 df["Worst Tut"] = worst_tuts
 columns.append("Worst Tut")
 df["Best 4 Tuts"] = Best_4_Tuts
 columns.append("Best 4 Tuts")

# best 4 and worst tut columns added
def Find_Total():
   Total = []
   for i in range(rows):
    total = 0
    for exam in columns:
       if(exam != "ID" and exam != "Worst Tut" and exam != "Best 4 Tuts"):
          total += int(df[exam][i])
    Total.append(total)
   df["Total"] = Total
   columns.append("Total")      

replace_woof()
Best_4()
Find_Total()

#print(df.to_string())

#plotting data of exams
#plotting starts from here

exams = [] # stores the exams 

x_exams = [] # on x-axis is the marks for each subject these lists are unsorted
y_exams = [] # on y-axis is the number of students who got the unique marks for that subject unsorted
sorted_x = [] # stores the sorted values of unique scores

marks = [] # this stores the marks for each inidividual exam

N_bins = 20 # number of bins
fig_width = 20
fig_height = 10

line_width = 2
labelfontsize = 15
titlefontsize = 15
textfontsize =15

legendfontsize = 12
barfontsize = 12

standard_dev = [] # stores the standard deviation of all the evaluatives
means = [] # stores the mean score of all the evaluatives
medians = [] # store the median score of all the evaluatives
modes = [] # store the modal acore of all the evaluatives
minumin_marks = []
maximum_marks = []
quantiles = []
skews = [] # stores the skew values of each evaluative

for col in columns:
   if(col == "ID" or col == "Best 4 Tuts" or col == "Worst Tut"):
        continue;
   exams.append(col)

def type_conversion():
 for col in exams:
    df[[col]] = df[[col]].apply(pd.to_numeric)

def prepare_plt_dataset():
   for exam in exams:
     
     mlist = []
     xlist = []
     ylist = []
    
     for j in range(rows):
         
         x = int(df[exam][j])
         mlist.append(x)
         
         if(x not in xlist):
             xlist.append(x) # stores each unique marks obtained in this test
             ylist.append(df.loc[df[exam] == x].shape[0]) # stores the list of the number of students for each unique marks
     
     marks.append(mlist)
     
     x_exams.append(xlist)
     y_exams.append(ylist)
     sorted_x.append(sorted(xlist))

def plot_data_line():
 
 style.use("ggplot")
 
 for i in range(len(exams)):
    pyplt.figure(figsize=(fig_width,fig_height))
    pyplt.plot(x_exams[i],y_exams[i], label = exams[i])
    pyplt.grid(True)
    pyplt.title("Line chart of number of students v/s marks obtained for {0}".format(exams[i]))
    pyplt.xlabel("Marks obtained")
    pyplt.ylabel("Number of students")
    pyplt.legend()  

 pyplt.tight_layout()
 pyplt.show()
 
type_conversion() # converts all integer data to int64 data type
prepare_plt_dataset() # prepares list of x and y values for plotting
plot_data_line() # creates line plot of prepared data

def mean(nlist):  # wanted to write own mean function
   sum = 0
   for x in nlist:
     sum += x
   return sum/len(nlist)

def median(nlist): #wanted to write own median function
 slist = sorted(nlist)
 n = len(nlist)
 if(n % 2 == 0):
    return (slist[n//2] + slist[n//2-1])/2
 else:
    return  slist[n//2]

def is_nan(x): # check if x is Nan
   return x != x

def plot_histogram():
   
   standard_dev = df[exams].std()
   modes  = df[exams].mode(numeric_only=True)
   minumin_marks = df[exams].min()
   maximum_marks = df[exams].max()
   quantiles = df[exams].quantile(q=[0.25,0.75])

   #fig,axs = pyplt.subplots(len(exams),1,figsize=(100,100))
   
   for i in range(len(exams)):
    
    pyplt.figure(figsize=(fig_width,fig_height))
    #axs.set_facecolor(colors.BASE_COLORS["k"]
    #fig.tight_layout(pad = 5)
    pyplt.grid(color = "grey", linestyle = "-",linewidth = 0.5,alpha = 0.6)
    N,bins,patches = pyplt.hist(marks[i],bins = N_bins)
    pyplt.xticks(bins)

    means.append(round(mean(marks[i]),2))
    medians.append(round(median(marks[i]),2))
    skews.append(skew(marks[i],axis = 0,bias = True))

    pyplt.axvline(means[i], color = "cyan",linestyle='dashed', linewidth=2, label = "Mean marks = {0}".format(means[i]))
    pyplt.axvline(medians[i], color = "blue",linestyle='dashed', linewidth=2, label = "Median marks = {0}".format(medians[i]))
    pyplt.axvline(minumin_marks[exams[i]], color = "red",linestyle='dashed', linewidth = line_width, label = "Minimum marks = {0}".format(minumin_marks[exams[i]]))
    pyplt.axvline(maximum_marks[exams[i]], color = "green",linestyle='dashed', linewidth = line_width, label = "Maximum marks = {0}".format(maximum_marks[exams[i]]))
    pyplt.axvline(quantiles[exams[i]][0.25], color = "brown",linestyle='dashed', linewidth = line_width, label = "First Quartile = {0}".format(quantiles[exams[i]][0.25]))
    pyplt.axvline(quantiles[exams[i]][0.75], color = "black",linestyle='dashed', linewidth = line_width, label = "Third Quartile = {0}".format(quantiles[exams[i]][0.75]))
    #multiple modes can be present
    for mode in modes[exams[i]]:
       if(not is_nan(mode)): # checking for Nan values
        pyplt.axvline(mode, color = "violet",linestyle='dashed', linewidth=1, label = "Modal marks = {0}".format(mode)) 
   
    fracs = ((N**(0.2))/max(N))
    norm = colors.Normalize(min(fracs),max(fracs))

    for frac,patch in zip(fracs,patches):
       color = pyplt.cm.viridis(norm(frac))
       patch.set_facecolor(color)

    pyplt.bar_label(patches,fontsize = barfontsize, color = "navy")
    pyplt.xlabel("Marks obtained",fontsize = labelfontsize)
    pyplt.ylabel("Number of students", fontsize = labelfontsize)
    pyplt.title("Frequency distribution for the data of {0}".format(exams[i]),fontsize = titlefontsize)
    pyplt.legend(title = "Legend",fontsize = legendfontsize)
    
    #adding relevant text
    pyplt.text(0,-15,"Standard Deviation = {0}".format(round(standard_dev[exams[i]],2)),fontsize = textfontsize)
    pyplt.text(0,-25, "Interquartile Range = {0}".format(round(quantiles[exams[i]][0.75]-quantiles[exams[i]][0.25],2)),fontsize = textfontsize)
    
    if(skews[0] > 0):    
     pyplt.text(0,-20,"The histogram is right-skewed",fontsize = textfontsize)
    elif(skews[0] < 0):
     pyplt.text(0,-20,"The histogram is left-skewed",fontsize = textfontsize)
    else:
     pyplt.text(0,-20,"The histogram is normally distributed", fontsize = textfontsize)
   
   pyplt.tight_layout()
   pyplt.show()

plot_histogram()

def conclude():
  print("The conclusion is as follows :-")
  conclusion = """\t\t  1.A high standard deviation value for an evaluative test
                    along with a low modal score implies that most students find
                    it harder get an average score on that evaluative. The given histogram plots 
                    clearly show that standard deviation is much lower in case of Tut tests 
                    as compared to compre and midsem. This clearly indicates that Tut tests are
                    easier to get a good grade in than midsem and compre.
                  2.In all evaluatives and the trend of the total score the difference in the
                    median and mean scores is very less indicating that the graph almost follows
                    a normal distribution with most of the students scoring in the range between 
                    the median and mode. But in all cases the modal score is very low indicating 
                    that a huge number of students are were unprepared for the evaluatives while
                    the others found it doable to get an avergage grade
                  3. Among Tut tests T1 was the hardest with the lowest mean score
                  4. Among compre A and compre B compre A was harder with a lower mean score"""
  print(conclusion)

conclude()
```
